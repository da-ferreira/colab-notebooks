```python
anos = [1980, 1982, 1984, 1986, 1988, 1990, 1992, 1994, 1996, 1998, 2000, 2002, 2004, 2006, 2008, 2010 , 2012]
co2_pais_a = [338.7, 341.2, 344.4, 347.2, 351.5, 354.2, 356.3, 358.6, 362.4, 366.5, 369.4, 373.2, 377.5, 381.9, 385.6, 389.9, 393.8]
co2_pais_b = [328.7, 331.2, 334.4, 337.2, 341.5, 344.2, 346.3, 348.6, 352.4, 356.5, 359.4, 363.2, 367.5, 371.9, 375.6, 379.9, 383.8]

plt.figure(figsize=(10, 8))

plt.plot(anos, co2_pais_a, 'b')  # Se trocar o '.' pelo 'o' a bolinha fica mais grossa
plt.plot(anos, co2_pais_b, 'r')


plt.title("Concentração de CO2 em partes por milhão (ppm)\n na atmosfera, de dois paises, entre 1980 e 2012\n")
plt.xlabel("Anos")
plt.ylabel("PPM")

plt.grid()
```
