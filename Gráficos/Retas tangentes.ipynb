```python
import matplotlib.pyplot as plt
import numpy as np
```

Sendo a curva y = x ^ 2, y = 2x - 1 é a equação da reta tangente ao ponto (1, 1) da curva 

```python
eixo_x = np.linspace(-9, 9, 100)
reta_tangente = np.linspace(-2, 4, 100)

plt.figure(figsize=(12, 10))

plt.plot(eixo_x, [y ** 2 for y in eixo_x], 'm', label="Curva y = x^2")

plt.plot(reta_tangente, [2 * x - 1 for x in reta_tangente], 'r', label="Reta tangente ao ponto P")
plt.plot([1, 4.45], [1, 20], 'k--', label="Reta secante PQ")

plt.plot(1, 1, 'ko', label='Ponto P(1, 1)')
plt.plot(4.45, 20, 'yo', label='Ponto Q(1, 1)')

plt.legend(loc=3, fontsize='large')
plt.grid()
plt.show()
```
