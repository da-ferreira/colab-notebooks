**A derivada f'(x) de uma função f(x) também é uma função.**

Exemplo 1: A função f(x) = x ^ 3 - x tem sua derivada f'(x) = 3x ^ 2 - 1



```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def funcao_f(x):
    return x ** 3 - x

def derivada_f(x):
    return 3 * x ** 2 - 1
```

```python
eixo_x = np.linspace(-10, 10, 100)

plt.figure(figsize=(12, 10))

plt.plot(eixo_x, [funcao_f(x) for x in eixo_x], 'b', label="f(x) = x^3 - x")
plt.plot(eixo_x, [derivada_f(x) for x in eixo_x], 'r', label="f'(x) = 3x^2 - 1")

plt.legend(loc=8, fontsize='x-large')
plt.grid()
plt.show()
```

Exemplo 2: A função f(x) = sqrt(x) tem sua derivada f'(x) = 1 / (2sqrt(x))

```python
def funcao_f2(x):
    return np.sqrt(x)

def derivada_f2(x):
    return 1 / (2 * np.sqrt(x))
```

```python
eixo_x = np.linspace(1, 25, 100)

plt.figure(figsize=(12, 10))

plt.plot(eixo_x, [funcao_f2(x) for x in eixo_x], 'b', label="f(x) = sqrt(x)")
plt.plot(eixo_x, [derivada_f2(x) for x in eixo_x], 'r', label="f'(x) = 1 / (2sqrt(x))")

plt.legend(fontsize='x-large')
plt.grid()
plt.show()
```
