```python
import matplotlib.pyplot as plt
```

```python
paises = ['Rússia', 'Canadá', 'EUA', 'China', 'Brasil']
extensao = [17098242, 9984670, 9833517, 9596960, 8515770]

plt.bar(paises, extensao, color='red')

plt.xlabel("Países")
plt.ylabel("Extensão")

plt.title("Os 5 maiores países do mundo em KM²")

plt.show()

# https://medium.com/horadecodar/gráficos-de-barra-com-matplotlib-85628bfc4351
```
