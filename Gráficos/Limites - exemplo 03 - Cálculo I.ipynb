```python
def lim_calculo(x):
    return (x**2 + x - 6) / (x - 2)
```

```python
x_menos = [0, 0.2, 0.5, 0.8, 1, 1.2, 1.5, 1.8, 1.9, 1.99, 1.999, 1.9999]   # Indo para 2 com valores menores que 2
x_mais = [4, 3.8, 3.5, 3.2, 3, 2.8, 2.5, 2.2, 2.1, 2.01, 2.001, 2.0001]    # Indo para 2 com valores maiores que 2

limite_x_menos = [lim_calculo(x) for x in x_menos]
limite_x_mais = [lim_calculo(x) for x in x_mais]

plt.style.use('default')

plt.figure(figsize=(10, 8))

plt.plot(x_menos, limite_x_menos, 'y', label="valores de x: 0, 0.2, 0.5, 0.8, 1, 1.2, 1.5, 1.8, 1.9, 1.99, 1.999, 1.9999")
plt.plot(x_mais, limite_x_mais, 'k', label="valores de x: 4, 3.8, 3.5, 3.2, 3, 2.8, 2.5, 2.2, 2.1, 2.01, 2.001, 2.0001")

plt.title("lim x->2 (x**2 + x - 6) / (x - 2) = 5", fontsize=18)

plt.legend(loc=4, fontsize='large')
plt.grid()
plt.show()
```
