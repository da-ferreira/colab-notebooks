```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import math

sns.set_style('ticks')

plt.rcParams['axes.spines.left'] = True
plt.rcParams['axes.spines.right'] = False
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.bottom'] = True
```

```python
eixo_x = np.linspace(1, 25, 100)

plt.figure(figsize=(12, 10))

plt.xlim(1, 30)
plt.ylim(0, 30)

plt.plot(eixo_x, [1] * len(eixo_x))                   # O(1) Constante
plt.plot(eixo_x, [math.log2(n) for n in eixo_x])      # O(log n)
plt.plot(eixo_x, eixo_x)                              # O(n)
plt.plot(eixo_x, [n * math.log2(n) for n in eixo_x])  # O(n log n) -> Log na base 2
plt.plot(eixo_x, [n ** 2 for n in eixo_x])            # O(n^2) 

plt.text(25.3, 1, "O(1)", fontsize=14)
plt.text(25.3, 4.5, "O(log n)", fontsize=14)
plt.text(25.3, 25, "O(n)", fontsize=14)
plt.text(7.5, 20, "O(n log n)", fontsize=14)
plt.text(2, 26, "O(n^2)", fontsize=14)

plt.title("big-O\n", fontsize=26)
plt.xlabel("n (Quantidade de dados)", fontsize=16)
plt.ylabel("Número de operações", fontsize=16)

plt.show()
```
