```python
x_menos = [1.5, 1.9, 1.95, 1.99, 1.995, 1.999] 
x_mais = [2.5, 2.1, 2.05, 2.01, 2.005, 2.001]

limite_x_menos = [round((((x ** 2) - (2 * x)) / (x ** 2) - x - 2), 6) for x in x_menos]
limite_x_mais =  [round((((x ** 2) - (2 * x)) / (x ** 2) - x - 2), 6) for x in x_mais]

plt.figure(figsize=(10, 8))

plt.plot(2, -4, 'k.', label="x -> 2")
plt.plot(x_menos, limite_x_menos, 'r', label="Valores de x: 1.5, 1.9, 1.95, 1.99, 1.995, 1.999")
plt.plot(x_mais, limite_x_mais, 'b', label="Valores de x: 2.5, 2.1, 2.05, 2.01, 2.005, 2.001")

plt.xticks([1, 1.2, 1.4, 1.6, 1.8, 2, 2.2, 2.4, 2.6, 2.8, 3])
plt.yticks([-3, -3.2, -3.4, -3.6, -3.8, -4, -4.2, -4.4, -4.6, -4.8, -5])

plt.title("lim x->2: x^2 -2x / x^2-x-2", fontsize=18)

plt.grid()
plt.legend(loc=3, fontsize='x-large')
plt.show()
```
