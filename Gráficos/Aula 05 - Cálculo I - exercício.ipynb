```python
def posicao(t):
    return 2 * np.cos(t) + 3 * np.sin(t)

def velocidade(t):
    return -2 * np.sin(t) + 3 * np.cos(t)

def aceleracao(t):
    return -2 * np.cos(t) - 3 * np.sin(t)

eixo_x = np.linspace(0, np.pi, 100)

plt.figure(figsize=(12, 10))
plt.title("0 <= t <= pi", fontsize=18)

plt.plot(eixo_x, [posicao(t) for t in eixo_x], 'y', label="Fun√ß√£o posi√ß√£o: 2cos t + 3sen t (m)")
plt.plot(eixo_x, [velocidade(t) for t in eixo_x], 'r', label="Fun√ß√£o velocidade: -2sen t + 3cos t (m/s)")
plt.plot(eixo_x, [aceleracao(t) for t in eixo_x], 'b', label="Fun√ß√£o acelera√ß√£o: -2cos t - 3sen t (m/s^2)")

plt.legend(loc=0, fontsize='large')  #  xx-small, x-small, small, medium, large, x-large, xx-large
plt.grid(True)
plt.show()
```
