```python
from prettytable import PrettyTable  # Biblioteca para gerar tabelas
```

```python
def media_velocidade(t_mais, t_menos=1):
    return ((10 * t_mais - 1.86 * t_mais ** 2) - (10 * t_menos - 1.86 * t_menos ** 2)) / (t_mais - t_menos)
```

```python
intervalos = [[1, 2], [1, 1.5], [1, 1.1], [1, 1.01], [1, 1.001], [1, 1.0001]]
velocidades_media = [media_velocidade(t[1]) for t in intervalos]

table = PrettyTable(["Intervalo de tempo", "Velocidade mdia (m/s)"])

for i in range(len(intervalos)):
    table.add_row([f"{intervalos[i][0]} <= t <= {intervalos[i][1] :.5f}", f"{velocidades_media[i] :.14f}"])

print(table)
```
