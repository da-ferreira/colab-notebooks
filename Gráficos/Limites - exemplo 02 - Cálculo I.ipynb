```python
def lim_calculo(t):
    return (sqrt((t ** 2) + 9) - 3) / (t ** 2)
```

```python
# Limites laterais

t_menos = [-5, -4, -3, -2, -1, -0.5, -0.05, -0.01]  # Indo para 0 com valores menores que 0
t_mais = [5, 4, 3, 2, 1, 0.5, 0.05, 0.01]           # Indo para 0 com valores maiores que 0

limite_t_menos = [lim_calculo(abs(x)) for x in t_menos]  # Limite lateral t -> 0-. O valor desse será igual ao valor de t->0+, por causa da raíz
limite_t_mais = [lim_calculo(x) for x in t_mais]         # Limite lateral t -> 0+

plt.figure(figsize=(10, 8))

plt.plot(t_menos, limite_t_menos, 'r', label="Valores de t = -5, -4, -3, -2, -1, -0.5, -0.05, -0.01")
plt.plot(t_mais, limite_t_mais, 'b', label="Valores de t = 5, 4, 3, 2, 1, 0.5, 0.05, 0.01")
#plt.plot(0, 0.16666666666666666, 'k.')  # Ponto de encontro

plt.title("lim t->0 (sqrt(t^2 + 9) - 3) / t^2 = 1/6", fontsize=18)

plt.legend(fontsize='x-large')
plt.grid()
plt.show()
```
