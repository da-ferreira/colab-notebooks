```python
def velocidade_media(t_mais, t_menos):
    """ velocidade m√©dia = ((i+1 ^ 2 - 8i+1 + 18) - (i ^ 2 + 8i + 18)) / (ti + 1 - t) """
    return ((t_mais ** 2 - 8 * t_mais + 18) - (t_menos ** 2 - 8 * t_menos + 18)) / (t_mais - t_menos)

intervalos = [[3, 4], [3.5, 4], [4, 5], [4, 4.5]]
resultado = []

for intervalo in intervalos:
    resultado.append(velocidade_media(intervalo[1], intervalo[0]))

print(resultado)
```

```python
intervalos_t = [[4, x] for x in np.linspace(5, 4.0001, 50)]
resultado_t = []

for intervalo in intervalos_t:
    resultado_t.append(velocidade_media(intervalo[1], intervalo[0]))

print(resultado_t)
```

```python
def funcao_s(t):
    return t ** 2 - 8 * t + 18

eixo_x = np.linspace(2.5, 5.5, 100)

plt.figure(figsize=(12, 10))

plt.title("Curva s(t) com retas secantes e suas inclina√ß√µes (m)", fontsize=17)

plt.plot(eixo_x, [funcao_s(t) for t in eixo_x], 'b', label="Curva de s(t)")

plt.plot([3, 4], [funcao_s(3), funcao_s(4)], 'r', label="Reta secante (3, 4), m = -1")          # [3, 4],   m = -1.0
plt.plot([3.5, 4], [funcao_s(3.5), funcao_s(4)], 'g', label="Reta secante (3.5, 4), m = -0.5")  # [3.5, 4], m = -0.5
plt.plot([4, 5], [funcao_s(4), funcao_s(5)], 'c', label="Reta secante (4, 5), m = 1")           # [4, 5],   m = 1.0
plt.plot([4, 4.5], [funcao_s(4), funcao_s(4.5)], 'y', label="Reta secante (4, 4.5), m = 0.5")   # [4, 4.5], m = 0.5

plt.legend(loc=0, fontsize='x-large')
plt.grid()
plt.show()
```

Na curva s(t), sua equa√ß√£o da reta tangente √© y = 2 ao ponto (4, 2) com sua  
inclina√ß√£o (coeficiente angular) sendo 2t - 8 (nesse caso a inclina√ß√£o da 0).

```python
def funcao_s(t):
    return t ** 2 - 8 * t + 18

eixo_x = np.linspace(3, 5, 100)

plt.figure(figsize=(12, 10))

plt.title("Curva s(t) com retas secantes e suas inclina√ß√µes (m)", fontsize=17)

plt.plot(eixo_x, [funcao_s(t) for t in eixo_x], 'b', label="Curva de s(t)")

plt.plot([3, 4], [funcao_s(3), funcao_s(4)], 'r', label="Reta secante (3, 4), m = -1")          # [3, 4],   m = -1.0
plt.plot([3.5, 4], [funcao_s(3.5), funcao_s(4)], 'g', label="Reta secante (3.5, 4), m = -0.5")  # [3.5, 4], m = -0.5
plt.plot([4, 5], [funcao_s(4), funcao_s(5)], 'c', label="Reta secante (4, 5), m = 1")           # [4, 5],   m = 1.0
plt.plot([4, 4.5], [funcao_s(4), funcao_s(4.5)], 'y', label="Reta secante (4, 4.5), m = 0.5")   # [4, 4.5], m = 0.5

reta_tangente = np.linspace(3.5, 4.5, 100)
plt.plot(reta_tangente, [2] * 100, 'k', label="Reta tangente ao ponto (4, 2)")
plt.plot(4, 2, 'ko', label="Ponto (4, 2)")

plt.legend(loc=0, fontsize='x-large')
plt.grid()
plt.show()
```
