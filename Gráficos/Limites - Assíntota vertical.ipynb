```python
# A reta x = 2 √© uma ass√≠ntota vertical de y = 1 / (x-2) ^ 2
# lim x->2 (1 / (x - 2) ^ 2) = +- infinito

def curva(x):
    return 1 / (x - 2) ** 2

x_menos = np.linspace(-5, 1.9, 100)  # 100 valores de -5 a 1.9, com igual diferen√ßa entre ele 
x_mais = np.linspace(9, 2.1, 100)

plt.figure(figsize=(10, 8))

plt.xlim(-6, 10)
plt.ylim(-1, 5)

plt.plot([2] * 7, [-0.95, 0, 1, 2, 3, 4, 5], 'b--')  # Reta x = 2
plt.plot(x_menos, [curva(x) for x in x_menos], 'k')
plt.plot(x_mais, [curva(x) for x in x_mais], 'k')

plt.text(2.1, 0.1, 'x = 2', fontsize=16)
plt.text(-0.5, 3.2, 'x -> 2+', fontsize=16)
plt.text(2.8, 3.2, 'x -> 2-', fontsize=16)

plt.grid()
plt.show()
```
