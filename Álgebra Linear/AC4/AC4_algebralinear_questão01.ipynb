```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```

def plot_polygon(polygon, newfigure=True, args={}, size=(15, 9)):
    x = np.array([v[0] for v in polygon])  # coordenada x
    y = np.array([v[1] for v in polygon])  # coordenada y

    if newfigure:
        plt.figure(figsize=size)
    
    plt.plot(x, y, **args)  # passando os argumentos para a funcao plot
```

# Questão 1: $(1,5 \; pontos)$

##Quais as coordenadas de um triângulo equilátero com lado de  1  unidade que começa centrado no ponto  (−1,0)  após ser translatado 5 vezes com o vetor  [2,−1]  e 4 vezes com o vetor  [4,5] ?

Apresente uma simulação apresentando cada translação considerando as translações com o vetor  [2,−1]  seguidas pelas translações com o vetor  [4,5] .

Apresente uma segunda simulação apresentando cada translação considerando as translações com o vetor  [2,−1]  intercaladas pelas translações com o vetor  [4,5] .

```

triangulo = [
        np.array([-1.5, -0.5, 1]),
        np.array([-0.5, -0.5, 1]),
        np.array([-1, 0.5, 1]),
        np.array([-1.5, -0.5, 1])
]

primeiro_deslocamento = [
        np.array([1, 0, 0]),
        np.array([0, 1, 0]),
        np.array([2, -1, 1])  # o x e y representa o vetor deslocamento.
]                         

plot_polygon(triangulo, args={'label': 'Posição inicial do triângulo', 'c': 'k'})

for i in range(5):
    triangulo = np.matmul(triangulo, primeiro_deslocamento)
    
    if i == 0:
        plot_polygon(triangulo, newfigure=False, args={'c': 'y', 'label': 'Deslocando com o vetor [2, -1]'})
    else:
        plot_polygon(triangulo, newfigure=False, args={'c': 'y'})


segundo_deslocamento = [
        np.array([1, 0, 0]),
        np.array([0, 1, 0]),
        np.array([4, 5, 1])  # o x e y representa o vetor deslocamento.
]                         

for i in range(4):
    triangulo = np.matmul(triangulo, segundo_deslocamento)

    if i == 0:
        plot_polygon(triangulo, newfigure=False, args={'c': 'b', 'label': 'Deslocando com o vetor [4, 5]'})

    elif i == 3:
        plot_polygon(triangulo, newfigure=False, args={'c': 'g', 'label': 'Posição final do triângulo'})
        
    else:
        plot_polygon(triangulo, newfigure=False, args={'c': 'b'})

plt.title('Simulação apresentando cada translação considerando as translações com o vetor [2,−1] seguidas pelas translações com o vetor [4,5]')

plt.legend(prop={'size': 15})
plt.tight_layout()
plt.grid();
```

```
triangulo = [
        np.array([-1.5, -0.5, 1]),
        np.array([-0.5, -0.5, 1]),
        np.array([-1, 0.5, 1]),
        np.array([-1.5, -0.5, 1])
]

primeiro_deslocamento = [
        np.array([1, 0, 0]),
        np.array([0, 1, 0]),
        np.array([2, -1, 1])  # o x e y representa o vetor deslocamento.
]                         

segundo_deslocamento = [
        np.array([1, 0, 0]),
        np.array([0, 1, 0]),
        np.array([4, 5, 1])  # o x e y representa o vetor deslocamento.
]                         


plot_polygon(triangulo, args={'label': 'Posição inicial do triângulo', 'c': 'k'})

for i in range(9):
    if i % 2 == 0:        
        triangulo = np.matmul(triangulo, primeiro_deslocamento)

        if i == 0:
            plot_polygon(triangulo, newfigure=False, args={'c': 'y', 'label': 'Deslocando com o vetor [2, -1]'})

        elif i == 8:
            plot_polygon(triangulo, newfigure=False, args={'c': 'g', 'label': 'Posição final do triângulo'})

        else:
            plot_polygon(triangulo, newfigure=False, args={'c': 'y'})
    else:
        triangulo = np.matmul(triangulo, segundo_deslocamento)

        if i == 1:
            plot_polygon(triangulo, newfigure=False, args={'c': 'b', 'label': 'Deslocando com o vetor [4, 5]'})
        else:
            plot_polygon(triangulo, newfigure=False, args={'c': 'b'})

plt.title('Simulação apresentando cada translação considerando as translações com o vetor [2,−1] intercaladas pelas translações com o vetor [4,5]')

plt.legend(prop={'size': 15})
plt.tight_layout()
plt.grid();
```
