```

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

t = np.linspace(0, 2 * np.pi, 100)  # variável muda

# -= Circulo central (rosto):

x1 = 8 * np.cos(t) + 0  # (Xc = 0, raio = 8)
y1 = 8 * np.sin(t) + 0  # (Yc = 0, raio = 8)

# -= Nariz (círculo)

x2 = 1.4 * np.cos(t) + 0  # (Xc = 0, raio = 1.4)
y2 = 1.4 * np.sin(t) + 0  # (Yc = 0, raio = 1.4)

# -= Chapéu (segmentos de retas)

# reta 1
angulo1 = np.radians(45)

reta_x1 = 0 + np.linspace(0, 6 * np.cos(angulo1), 100)  # coordenadas de inicio do segmento = (0, 8) 
reta_y1 = 8 + np.linspace(0, 6 * np.sin(angulo1), 100)  # 0 a 6 é o tamanho do segmento.

# reta 2
angulo2 = np.radians(104)

reta_x2 = 6 + np.linspace(0, 7.15 * np.cos(angulo2), 100)  # coordenadas de inicio do segmento = (6, 5.30) 
reta_y2 = 5.30 + np.linspace(0, 7.15 * np.sin(angulo2), 100)  # 0 a 7.15 é o tamanho do segmento.  

# -= Boca (parábolas)

# parte de baixo
t2 = np.linspace(-6, 6, 100)  # pontas da parabola (eixo x) de 6 a -6

x3 = t2
y3 = 0.2 * (t2 - 0) ** 2 + -6.5  # (abertura = 0.2), Xvertice = 0, Yvertice = -6.5

# parte de cima
x4 = t2
y4 = 0.134 * (t2 - 0) ** 2 + -4.1  # (abertura = 0.134), Xvertice = 0, Yvertice = -4.1

# -= Lingua (parábola) 

t3 = np.linspace(-1.3, 1.3, 100)  # pontas da parabola (eixo x) de -1.3 a 1.3

x5 = t3
y5 = -0.85 * (t3 - 0) ** 2 + -4.7  # (abertura = -0.85), Xvertice = 0, Yvertice = -4.7

# -= Olhos (hipérbole horizontal)

t4 = np.linspace(-2.8, 2.8, 100)  # Intervalo da hipérbole

abertura_a, abertura_b = 0.35, 0.35  # abertura
xc, yc = 0, 3.5  # Centro x e y

x_direito = abertura_a * np.cosh(t4) + xc
x_esquerdo = -abertura_b * np.cosh(t4) + xc  # espelhando o eixo x
y_hiperbole = abertura_b * np.sinh(t4) + yc

# A figura:

plt.figure(figsize=(15, 15))  # Tamanho da figura

plt.plot(x1, y1, 'b')  # círculo grande ('b' = cor azul)

plt.plot(x2, y2, 'b')  # nariz

plt.plot(reta_x1, reta_y1, 'b')  # reta 1 do chapéu 
plt.plot(reta_x2, reta_y2, 'b')  # reta 2 do chapéu 

plt.plot(x3, y3, 'b')  # boca (parte de baixo) 
plt.plot(x4, y4, 'b')  # boca (parte de cima) 

plt.plot(x5, y5, 'b')  # lingua

plt.plot(x_direito, y_hiperbole, 'b')  # Olho direito
plt.plot(x_esquerdo, y_hiperbole, 'b')  # Olho esquerdo

#plt.grid()  # grades
```
