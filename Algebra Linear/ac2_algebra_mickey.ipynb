```

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

t = np.linspace(0, 2 * np.pi, 100)  # variável muda

# -= Circulo central (rosto):

x1 = 8 * np.cos(t) + 0  # (Xc = 0, raio = 8)
y1 = 8 * np.sin(t) + 0  # (Yc = 0, raio = 8)

# -= Orelha esquerda e direita:

r = 3  # <- Raio da orelha esquerda e direita

x2 = r * np.cos(t) + -7.8  # (Xc = -7.8) esquerda 
y2 = r * np.sin(t) + 7.8  # (Yc = 7.8)

x3 = r * np.cos(t) + 7.8  # (Xc = 7.8) direita
y3 = r * np.sin(t) + 7.8  # (Yc = 7.8)

# -= Olho esquerdo e direito (elipse):

a, b = 2, -1.2  # tamanho da elipse ( (sua área))

x4 = b * np.cos(t) + -4  # (Xc = -4) esquerdo
y4 = a * np.sin(t) + 2  # (Yc = 2)

x5 = b * np.cos(t) + 4  # (Xc = 4) direito
y5 = a * np.sin(t) + 2  # (Yc = 2)

# -= Bolinha do olho esquerdo e direito:

x6 = 0.8 * np.cos(t) + -4  # (Xc = -4, raio = 0.8) esquerdo
y6 = 0.8 * np.sin(t) + 1.5  # (Yc = 1.5, raio = 0.8)

x7 = 0.8 * np.cos(t) + 4  # (Xc = 4, raio = 0.8) direito
y7 = 0.8 * np.sin(t) + 1.5  # (Yc = 1.5, raio = 0.8)

# -= Boca (parábola)

# parte de cima
t2 = np.linspace(-5.5, 5.5, 100)  # pontas da parabola (eixo x) de -5.5 a 5.5

x8 = t2
y8 = 0.06 * (t2 - 0) ** 2 + -4  # (abertura = 0.06), Xvertice = 0, Yvertice = -4

# parte de baixo
t3 = np.linspace(-2.5, 2.5, 100)  # pontas da parabola (eixo x) de -2.5 a 2.5

x9 = t3
y9 = 0.24 * (t3 - 0) ** 2 + -5.15  # (abertura = 0.24), Xvertice = 0, Yvertice = -5.15

# -= Nariz (parábola e elipse)

# parte de baixo, elipse
a2, b2 = 1.4, 0.7  # tamanho da elipse (sua área)

x10 = a2 * np.cos(t) + 0  # (Xc = 0)
y10 = b2 * np.sin(t) + -0.7  # (Yc = -0.7)

# parte de cima, parabola
t4 = np.linspace(-3.5, 3.5, 100)  # pontas da parabola (eixo x) de -3.5 a 3.5

x11 = t4
y11 = -0.08 * (t4 - 0) ** 2 + 0.2  # (abertura = -0.08), Xvertice = 0, Yvertice = 0.2

# A figura:

plt.figure(figsize=(10, 10))  # <- Tamanho da figura

plt.plot(x1, y1, 'k')  # circulo grande ('k' = cor preta)

plt.plot(x2, y2, 'k')  # orelha esquerda 
plt.plot(x3, y3, 'k')  # orelha direita

plt.plot(x4, y4, 'k')  # olho esquerdo
plt.plot(x5, y5, 'k')  # olho direito
plt.plot(x6, y6, 'k')  # bolinha do olho esquerdo
plt.plot(x7, y7, 'k')  # bolinha do olho direito

plt.plot(x8, y8, 'k')  # boca (parte de cima) 
plt.plot(x9, y9, 'k')  # boca (parte de baixo) 

plt.plot(x10, y10, 'k')  # nariz (parte de baixo) 
plt.plot(x11, y11, 'k')  # nariz (parte de cima) 
```
