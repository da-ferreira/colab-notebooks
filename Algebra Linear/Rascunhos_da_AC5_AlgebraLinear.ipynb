```
import numpy as np
import matplotlib.pyplot as plt
```

```

def plot_polygon(polygon, newfigure=True, args={}, size=(15, 9), grades=False):
    x = np.array([v[0] for v in polygon])  # coordenada x
    y = np.array([v[1] for v in polygon])  # coordenada y

    if newfigure:
        plt.figure(figsize=size)

    plt.plot(x, y, **args)  # passando os argumentos para a funcao plot
    plt.grid(grades)


# Formula para achar a area de qualquer poligono a partir de seus pontos
def shoelace_formula(vertices):
    soma1 = 0
    soma2 = 0

    for i in range(len(vertices) - 1):
        soma1 += vertices[i][0] * vertices[i + 1][1]
        soma2 += vertices[i][1] * vertices[i + 1][0]
    
    soma1 += vertices[-1][0] * vertices[0][1]
    soma2 += vertices[-1][1] * vertices[0][0]

    area = abs(soma1 - soma2)
    area = area / 2

    return area
```

# Demonstração de Elipse (círculo alongado)

```

t = np.linspace(0, 2 * np.pi) 

# Onde 3 e 4 são os raios, 3 na horizontal e 4 na vertical
# E -2 e 3 é o centro, Xc = -2, Yc = 3

x = 3 * np.cos(t) - 2
y = 4 * np.sin(t) + 3

plt.tight_layout()

plt.plot(x, y)
plt.grid();
```

# Qual o ângulo entre os vetores [2, -3, 1, 5] e [6, 5, -7, 2]?

```

vetor_v = np.array([2, -3, 1, 5])
vetor_u = np.array([6, 5, -7, 2])

produto_interno = np.dot(vetor_v, vetor_u)
norma_v = np.linalg.norm(vetor_v)
norma_u = np.linalg.norm(vetor_u)

angulo = np.arccos(produto_interno / (norma_v * norma_u))
print(np.degrees(angulo))
```

## Se A = [[3, 0], [-1, -1]], B = [[-2, 1], [5, -2]] e C = [[1, 0], [0, 1]], qual o determinante de: A * B + C? 

```

a = np.array([[3,   0],
              [-1, -1]])

b = np.array([[-2,  1],
              [5,  -2]])

c = np.array([[1, 0],
              [0, 1]])

multiplicacao = np.matmul(a, b)
soma = multiplicacao + c

print(np.linalg.det(soma))
```

##Qual o determinante desta matriz [[3, 0, -1], [-2, 1, 1], [0, -3, 2]]?

```

matriz = np.array([
           [3, 0, -1],
           [-2, 1, 1],
           [0, -3, 2]        
])

print(np.linalg.det(matriz))
```

## Considere o polígono formado pelos pontos {(3, 4), (-3, 4), (-1, 1), (-3, -2), (3, -2), (1, 1)}. Qual opção contém a(s) matriz(es) que implementam um escalonamento que triplica a área do objeto e mantém seu centro na mesma posição?

```

poligono = np.array([
        [3,   4, 1],
        [-3,  4, 1],
        [-1,  1, 1], 
        [-3, -2, 1], 
        [3,  -2, 1], 
        [1,   1, 1],
        [3,   4, 1]    
])

centro = np.mean(poligono[:-1], axis=0)  # (0, 1)
area_original = shoelace_formula(poligono[:-1])  # 24

ida_ao_centro = np.array([
            [1,  0,  0],
            [0,  1,  0], 
            [0, -1,  1]
])

escalonamento = np.array([
            [6,  0,  0],
            [0, 0.5, 0], 
            [0,  0,  1]
])

volta_ao_centro = np.array([
            [1, 0, 0],
            [0, 1, 0], 
            [0, 1, 1]
])

poligono_escalonado = np.matmul(poligono, ida_ao_centro)

print(poligono_escalonado, '\n')

poligono_escalonado = np.matmul(poligono_escalonado, escalonamento)

print(poligono_escalonado, '\n')

poligono_escalonado = np.matmul(poligono_escalonado, volta_ao_centro)

print(poligono_escalonado, '\n')

area_escalonado = shoelace_formula(poligono_escalonado[:-1])
legenda_poligono_original = f'Poligono Original\nArea: {area_original}'
legenda_poligono_escalonado = f'Poligono Escalonado\nArea: {area_escalonado}'

plot_polygon(poligono, size=(8, 8), args={'label': f'{legenda_poligono_original}', 'c': 'y'}, grades=True)
plt.legend(prop={'size': 12}, loc=10)

plot_polygon(poligono_escalonado, size=(10, 10), args={'label': f'{legenda_poligono_escalonado}', 'c': 'b'}, grades=True)
plt.legend(prop={'size': 12}, loc=10)

```
