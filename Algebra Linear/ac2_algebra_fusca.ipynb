```

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

t = np.linspace(0, 2 * np.pi, 100)  # variável muda

# --- RODAS ---

# Esquerda
x_roda_esquerda = 1 * np.cos(t) + -5  # Xc = -5, raio = 1
y_roda_esquerda = 1 * np.sin(t) + -4  # Yc = -4, raio = 1

# bolinha da esquerda
x_bolinha_esquerda = 0.15 * np.cos(t) + -5  # Xc = -5, raio = 0.15
y_bolinha_esquerda = 0.15 * np.sin(t) + -4  # Xc = -4, raio = 0.15

# Direita

x_roda_direita = 1 * np.cos(t) + 2  # Xc = 2, raio = 1
y_roda_direita = 1 * np.sin(t) + -4  # Yc = -4, raio = 1

# bolinha da direita

x_bolinha_direita = 0.15 * np.cos(t) + 2  # Xc = 2, raio = 0.15
y_bolinha_direita = 0.15 * np.sin(t) + -4  # Xc = -4, raio = 0.15

# --- CHÃO ---

chao_x = 9 + np.linspace(0, 20 * np.cos(np.radians(180)), 100)  # angulo de 180
chao_y = -5 + np.linspace(0, 20 * np.sin(np.radians(180)), 100)

# --- LINHAS DO CARRO ---

linha_entre_roda_x = -4 + np.linspace(0, -5 * np.cos(np.radians(180)), 100)  # coordenadas de inicio x = -4, y = -4
linha_entre_roda_y = -4 + np.linspace(0, -5 * np.sin(np.radians(180)), 100)

linha_esquerda_x = -7 + np.linspace(0, -1 * np.cos(np.radians(180)), 100) 
linha_esquerda_y = -4 + np.linspace(0, -1 * np.sin(np.radians(180)), 100)

linha_direita_x = 3 + np.linspace(0, -2 * np.cos(np.radians(180)), 100) 
linha_direita_y = -4 + np.linspace(0, -2 * np.sin(np.radians(180)), 100)

# --- CARCAÇA ---

# Parte maior
t2 = np.linspace(-7, 0, 100)  # pontas da parabola (eixo x) de -7 a 0

carcaca_maior_x = t2
carcaca_maior_y = -0.25 * (t2 - -3) ** 2 + 0  # (abertura = -0.25), Xvertice = -3, Yvertice = 0

# Parte menor
t3 = np.linspace(0, 5, 100)  # pontas da parabola (eixo x) de 0 a 5

carcaca_menor_x = t3
carcaca_menor_y = -0.17 * (t3 - 1.5) ** 2 + -1.9  # (abertura = -0.17), Xvertice = 1.5, Yvertice = -1.9

# --- PARTE DA PORTA ---

linha_porta_x1 = -2.5 + np.linspace(0, 3.92 * np.cos(np.radians(90)), 100)  # angulo de 90 
linha_porta_y1 = -4 + np.linspace(0, 3.92 * np.sin(np.radians(90)), 100)  # tamanho: de 0 a 3.92

linha_porta_x2 = -6 + np.linspace(0, -6 * np.cos(np.radians(180)), 100)  # angulo de 180 
linha_porta_y2 = -2.27 + np.linspace(0, -6 * np.sin(np.radians(180)), 100)  # tamanho: de 0 a -6

# maçaneta (parábola)
t4 = np.linspace(-2.4, -2, 100)  # pontas da parabola (eixo x) de 0 a 5

macaneta_x = t4
macaneta_y = -1 * (t4 - -2.2) ** 2 + -2.4  # (abertura = -0.17), Xvertice = 1.5, Yvertice = -1.9

# --- LANTERNA (segmentos) ---

# linha de baixo
lanterna_x1 = 4.05 + np.linspace(0, 0.5 * np.cos(np.radians(180)), 100)   # angulo de 180
lanterna_y1 = -3 + np.linspace(0, 0.5 * np.sin(np.radians(180)), 100)  # tamanho: de 0 a 0.5

# linha de cima
lanterna_x2 = 3.80 + np.linspace(0, 0.25 * np.cos(np.radians(180)), 100)   # angulo de 180
lanterna_y2 = -2.8 + np.linspace(0, 0.25 * np.sin(np.radians(180)), 100)  # tamanho: de 0 a 0.25

# linha do meio
lanterna_x3 = 3.55 + np.linspace(0, -0.20 * np.cos(np.radians(90)), 100)   # angulo de 90
lanterna_y3 = -2.8 + np.linspace(0, -0.20 * np.sin(np.radians(90)), 100)  # tamanho: de 0 a -0.20

# Configurando a figura:

plt.figure(figsize=(30, 12))  # Tamanho da figura

plt.plot(x_roda_esquerda, y_roda_esquerda, 'k')  # Roda esquerda ('k' = cor preta)
plt.plot(x_roda_direita, y_roda_direita, 'k')  # Roda direita

plt.plot(chao_x, chao_y, 'k')  # chao

plt.plot(x_bolinha_esquerda, y_bolinha_esquerda, 'k')  # bolinha esquerda
plt.plot(x_bolinha_direita, y_bolinha_direita, 'k')  # bolinha direita

plt.plot(linha_entre_roda_x, linha_entre_roda_y, 'k') # linha do meio
plt.plot(linha_esquerda_x, linha_esquerda_y, 'k') # linha esquerda
plt.plot(linha_direita_x, linha_direita_y, 'k') # linha direita

plt.plot(carcaca_maior_x, carcaca_maior_y, 'k')  # Parte maior
plt.plot(carcaca_menor_x, carcaca_menor_y, 'k')  # Parte menor

plt.plot(linha_porta_x1, linha_porta_y1, 'k')  # linha vertical
plt.plot(linha_porta_x2, linha_porta_y2, 'k')  # linha horizontal

plt.plot(macaneta_x, macaneta_y, 'k')  # maçaneta

plt.plot(lanterna_x1, lanterna_y1, 'k')  # lanterna, linha de baixo
plt.plot(lanterna_x2, lanterna_y2, 'k')  # lanterna, linha de cima
plt.plot(lanterna_x3, lanterna_y3, 'k')  # lanterna, linha do meio
```
