```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```

def plot_polygon(polygon, newfigure=True, args={}, size=(15, 9)):
    x = np.array([v[0] for v in polygon])  # coordenada x
    y = np.array([v[1] for v in polygon])  # coordenada y

    if newfigure:
        plt.figure(figsize=size)
    
    plt.plot(x, y, **args)  # passando os argumentos para a funcao plot
```

# Questão 2: $(1,5 \; pontos)$

Qual o menor ângulo $\alpha$ que devemos rotacionar um pentágono regular em torno de seu centro para que cada vértice coincida com a posição do vértice adjacente?

Apresente uma simulação utilizando um polígono centrado no ponto $(5,5)$ e lado de $1$ unidade.

```

# RESPOSTA
# O menor ângulo que se deve rotacionar o pentágono em torno de seu centro
# de forma que o vértice coincida com o vértice adjacente é 72ª

t = np.linspace(0, 2 * np.pi, 6)

pentagono_x = 1 * np.cos(t) + 5
pentagono_y = 1 * np.sin(t) + 5

pentagono = np.vstack((
                    pentagono_x,
                    pentagono_y,
                    np.ones(6,))).T


plot_polygon(pentagono, size=(6, 6), args={'c': 'b', 'label': 'Pentagono posição inicial'})

plt.legend(prop={'size': 10})  # Tamanho da legenda
plt.tight_layout()
plt.grid()


angulo = np.deg2rad(72)  # Converte graus para radianos

rotacao = np.array([
        [np.cos(angulo), np.sin(angulo),  0],
        [-np.sin(angulo), np.cos(angulo), 0],
        [0,              0,               1]
])

# Transposta da matriz para que a rotação seja no sentido horário.
rotacao = rotacao.T 

# Como o Xc e Yc != (0, 0) é necessário voltar ao ponto 0, 0
# fazer a rotação e retornar o objeto aos pontos originais.

ida_ao_centro = np.array([
            [1,   0,  0],
            [0,   1,  0],
            [-5, -5,  1]
])

volta_do_centro = np.array([
            [1, 0, 0],
            [0, 1, 0],
            [5, 5, 1]
])


auxiliar = np.matmul(pentagono, ida_ao_centro)
auxiliar = np.matmul(auxiliar, rotacao)
pentagono_rotacionado = np.matmul(auxiliar, volta_do_centro)

plot_polygon(pentagono_rotacionado, size=(6, 6), args={'c': 'y', 'label': 'Pentagono rotacionado 72º'})

plt.tight_layout()
plt.legend(prop={'size': 10})  # Tamanho da legenda
plt.grid()

```
